```python
import re
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 1. Define the target Job Description (The Benchmark)
job_description = """
We are looking for a Remote Machine Learning Engineer with strong Python programming skills. 
The ideal candidate must have experience with Natural Language Processing (NLP), Scikit-Learn, 
and Pandas for handling structured data. Experience with SQL database management and building 
automated data pipelines is highly preferred.
"""

# 2. Create a simulated dataset of candidate resumes to screen
resumes_data = {
    "Candidate_Name": ["Amina Yusuf", "John Doe", "Elena Rostova", "Kwame Mensah"],
    "Resume_Text": [
        "Experienced Data Scientist specializing in Natural Language Processing (NLP) and Machine Learning. Highly proficient in Python, Scikit-Learn, and Pandas. Developed automated data pipelines and managed production SQL databases.",
        "Graphic Designer with 5 years of experience in Adobe Photoshop, Illustrator, and UI/UX design. Creative thinker with background in marketing layouts and frontend HTML/CSS website adjustments.",
        "Software Engineer with a focus on Python backend development. Skillful with Django and Flask web frameworks. Basic knowledge of SQL databases and data analysis using Pandas, looking to learn more about Machine Learning.",
        "Senior ML Engineer. Deep expertise in Python, Scikit-Learn, and building complex NLP models. Background includes advanced text processing, neural networks, and optimizing enterprise data pipelines."
    ]
}

# Convert to a clean tracking DataFrame
df_candidates = pd.DataFrame(resumes_data)
df_candidates
```

```python
# 1. Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# 2. Combine the Job Description and all candidate resumes into one list for vectorization
all_texts = [job_description] + df_candidates['Resume_Text'].tolist()

# 3. Transform the text data into numerical matrix vectors
tfidf_matrix = vectorizer.fit_transform(all_texts)

# 4. Separate the Job Description vector from the Resume vectors
jd_vector = tfidf_matrix[0]       # First element is our Job Description benchmark
resume_vectors = tfidf_matrix[1:] # The remaining elements are the candidate vectors

# 5. Calculate Cosine Similarity scores between the JD and each resume
similarity_scores = cosine_similarity(jd_vector, resume_vectors).flatten()

# 6. Append the calculated scores to our tracking DataFrame and format as percentages
df_candidates['Match_Score_%'] = np.round(similarity_scores * 100, 2)

# 7. Sort the candidates based on score from highest match to lowest match (Ranking)
df_ranked_candidates = df_candidates.sort_values(by='Match_Score_%', ascending=False).reset_index(drop=True)

# Print the final prioritized hiring dashboard
print("--- 📄 ATS CANDIDATE RANKING SYSTEM INITIALIZED ---")
df_ranked_candidates[['Candidate_Name', 'Match_Score_%']]
```

```python
# 1. Define the mandatory skills we are tracking for the position
required_skills = ["Python", "NLP", "Scikit-Learn", "Pandas", "SQL", "Pipelines"]

def analyze_skill_gaps(text, skill_list):
    # Standardize text to lowercase for exact keyword comparison
    text_lower = text.lower()
    found = [skill for skill in skill_list if skill.lower() in text_lower]
    missing = [skill for skill in skill_list if skill.lower() not in text_lower]
    return found, missing

# 2. Apply the extraction across all ranked candidates
found_list = []
missing_list = []

for idx, row in df_ranked_candidates.iterrows():
    found, missing = analyze_skill_gaps(row['Resume_Text'], required_skills)
    found_list.append(", ".join(found) if found else "None")
    missing_list.append(", ".join(missing) if missing else "None")

# 3. Append analysis to the dashboard
df_ranked_candidates['Skills_Matched'] = found_list
df_ranked_candidates['Missing_Required_Skills'] = missing_list

print("--- 🎯 FINAL ATS SCREENING & REPORT GENERATED ---")
df_ranked_candidates[['Candidate_Name', 'Match_Score_%', 'Skills_Matched', 'Missing_Required_Skills']]
```

## 📊 System Architecture & Methodology

This automated candidate screening engine operates across an optimization pipeline using advanced text space vectorization:

1. **Text Vectorization (`TF-IDF`):** Converts unstructured string texts into frequency-inverse document matrices, prioritizing rare terms like *NLP* or *Scikit-Learn* over common English stop words.
2. **Mathematical Ranking (`Cosine Similarity`):** Computes the angular similarity metric between the multi-dimensional Job Description baseline vector ($A$) and each applicant's Resume vector ($B$):
$$Similarity = \frac{A \cdot B}{\|A\| \|B\|}$$
3. **Skill Gap Logic Parsing:** A deterministic NLU scanner cross-references candidate inputs against baseline operational thresholds to dynamically flag explicit skill deficiencies.
